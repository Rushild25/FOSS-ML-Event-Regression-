To implement train test split on the refined dataset without using external libraries my approach is to shuffle/randomize the data and then split it into X_train and X_test (corresponding y_train and y_test).

```python
import random

def train_test_splitter(df, target_col = "price", test_size=0.2, seed=None):
  if seed is not None:
    random.seed(seed)

  df = df.sample(frac=1, random_state=seed).reset_index(drop=True)

  split_idx = int(len(df) * (1 - test_size))

  train_df = df.iloc[:split_idx]
  test_df = df.iloc[split_idx:]

  X_train = train_df.drop(columns=[target_col])
  y_train = train_df[target_col]

  X_test = test_df.drop(columns=[target_col])
  y_test = test_df[target_col]

  return X_train, X_test, y_train, y_test

df = pd.read_csv("updated_wine_data.csv")
X_train, X_test, y_train, y_test = train_test_splitter(df, test_size=0.2, seed=42)

print("X_train: \n", X_train)
print("y_train: \n", y_train)
print("X_test: \n", X_test)
print("y_test: \n", y_test)
```
